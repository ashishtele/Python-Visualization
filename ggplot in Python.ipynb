```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from ggplot import *
```

```python
data = pd.read_csv("E:/Python/Scripts/Data/airlines.csv")
```

```python
data.shape
```

```python
fig, ax = plt.subplots(figsize=(20,8))

# no. of delays vs no. of flights colored by year
ggplot(data[data.Delayed<10000], aes(x = 'Flights.Total', y='Delayed', color = 'Year')) +\
geom_point()
```

```python
ggplot(data, aes(x='Year', y='Carriers.Total')) +\
geom_violin(fill="#3366FF")
```

```python
ggplot(data, aes(x='# of Delays.Carrier',y='# of Delays.Weather'))+\
geom_point(aes(color='Year'))
```

```python
ggplot(data, aes(x='Month Name', y = 'Delayed', color='Month Name'))+\
geom_boxplot(aes(fill='Month Name'))
```
