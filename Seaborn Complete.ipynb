```python
# histograms, KDE, densities
data = np.random.multivariate_normal([0,0],[[5,2],[2,2]], size = 2000)
data = pd.DataFrame(data, columns=['x','y'])

for col in 'xy':
    plt.hist(data[col], normed=True, alpha=0.5)
```

```python
# kernel density estimation
for col in 'xy':
    sns.kdeplot(data[col], shade=True)
```

```python
# faceted histogram

tips = sns.load_dataset('tips')
tips.head()
```

```python
# marathon dataset

data = pd.read_csv('E:/Python/Scripts/Data/marathon.csv')
data.head()

rng = data['split'].str.split(':')
rng1 = data['final'].str.split(':')

a=[]
for i in range(0,len(rng)):
    a.append(int(rng[i][0])*3600+int(rng[i][1])*60+int(rng[i][2]))

b=[]
for i in range(0,len(rng)):
    b.append(int(rng1[i][0])*3600+int(rng1[i][1])*60+int(rng1[i][2]))

col = ['spl']
df = pd.DataFrame(np.array(a).reshape(len(a),1), columns=col)

col1 = ['fnl']
df1 = pd.DataFrame(np.array(b).reshape(len(b),1), columns=col1)

df2 = np.hstack((np.array(data),np.array(df)))
df3 = np.hstack((np.array(df2),np.array(df1)))
```

```python
# column names
colm = (np.append(np.append(data.columns.values,col),col1))
```

```python
df_fnl = pd.DataFrame(df3, columns = colm)
```
